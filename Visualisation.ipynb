# Visualization
In dit notebook zullen we de het passen die tijdens de wedstrijd hebben plaats gevonden. We zullen dit doen met behulp van de python visualization libarary Bokeh. Bokeh leverd hele mooie grafieken op en je kan er interactie aan toevoegen. 
<br />
<br />
We importeren een de benodige libraries.

```python
import pandas as pd

from bokeh.charts import Scatter, show
from bokeh.models import Jitter
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

We hebben in het notebook 'Data Collection & Munging' datasets gemaakt gebaseerd op de xml dataset. Deze datasets zullen we hier gebruiken. 

```python
df = pd.read_csv('datasets/bewerkte-datasets/Balbezit NEC-Roda.csv')

df = df.set_index('Unnamed: 0')
df.index.name = None
```

We maken hier het figuur aan. We voegen er een plaatje aan toe als achtergrond. En plotten elke paas als een stipje op het figuur. We kleuren de stipjes rood voor de tegenstander en blauw voor NEC. Dit geeft wat meer overzicht.

```python
p = figure( x_range=(0.3,3.7), y_range=(0,4.5), width=437,height=647,title="Doelpunten")

p.image_url(url=['voetbalveld-indeling.png'], x=2, y=2.5, w=2.8, h=4.0, anchor="center")

steps = 2
doelpuntNummer = 1

goal = df[df['code'].str.contains('Goal')].index.tolist()[doelpuntNummer]

firstTenX = df.iloc[goal-steps:goal].x
firstTenY = df.iloc[goal-steps:goal].y
firstTenBezit = df.iloc[goal-steps:goal].bezit

for i, x in enumerate(firstTenX):
    y = firstTenY.iloc[i]
    club = 'TEG' if firstTenBezit.iloc[i] == ' TEG' else 'NEC'
    color = 'red' if club == 'TEG' else 'blue'
    p.circle(x={'value': x, 'transform': Jitter(width=0.4)}, y=y, color=color, size=10, legend=club)
    
p.legend.location = "bottom_left"
show(p)
```
